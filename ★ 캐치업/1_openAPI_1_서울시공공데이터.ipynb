```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# API ê¸°ë³¸ ì„¤ì •
base_url = "http://openapi.seoul.go.kr:8088"  # ê¸°ë³¸ URL
api_key = "4b46504e7762616b3130304c66566247"  # ì„œìš¸ì‹œì—ì„œ ì¸ì¦í‚¤ë¥¼ ë°›ìŒ
response_format = "json"  # ì‘ë‹µ í˜•ì‹
service_name = "SeoulLibraryLectureInfo"  # API ì„œë¹„ìŠ¤ ì´ë¦„
start_index = 1  # ì‹œì‘ ë°ì´í„° ë²ˆí˜¸
end_index = 370  # ë ë°ì´í„° ë²ˆí˜¸ (ì œê³µ ë°ì´í„° ê°¯ìˆ˜ì— ë”°ë¼ ë‹¤ë¦„)

# URL ì¡°í•©
url = f"{base_url}/{api_key}/{response_format}/{service_name}/{start_index}/{end_index}"
print(f"Requesting URL: {url}")

# API í˜¸ì¶œ
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    libData = data.get(service_name, {}).get("row", [])
else:
    print(f"Failed to fetch data. Status Code: {response.status_code}")
    libData = []

# CONTENT íƒœê·¸ë¥¼ ì œê±°í•˜ë©° ë°ì´í„°í”„ë ˆì„ ìƒì„±
processed_data = []
for entry in libData:
    # HTML íƒœê·¸ ì œê±°
    if 'CONTENT' in entry:
        soup = BeautifulSoup(entry['CONTENT'], "html.parser")
        entry['CONTENT'] = soup.get_text(separator="\n").strip()  # HTML íƒœê·¸ ì œê±°
    processed_data.append(entry)

# ë°ì´í„°í”„ë ˆì„ ìƒì„±
df = pd.DataFrame(processed_data)

# í•„ìš”í•œ ì—´ ì„ íƒ
result = df[['LECTURE_ID', 'TITLE', 'TERM_START_TIME', 'CONTENT']]

# ê²°ê³¼ ì¶œë ¥
print(result)
```
