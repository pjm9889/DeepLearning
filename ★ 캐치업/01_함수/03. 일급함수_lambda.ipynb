## [1] 일급함수란_람다 함수(lambda function)
- pdf 34페이지
- 람다함수: 이름이 없는 익명함수( f조차 없는 함수: “입력 x를 받아 x+1을 반환하는 함수”를 기호로만 표현)
    - 람다함수는 파이썬에서 이름이 없는 함수(익명 함수)

- 일급함수의 개념(First-class function): 제일 좋은 대우를 받는 함수, 일등석(First Class)’처럼, 특별히 제한 없이 모든 권리를 가진 존재
    - 옛날 프로그래밍 언어(C, Fortran 등)에서는 함수는 ‘명령’처럼만 존재, 즉, “호출할 수는 있지만”, “변수에 담거나 전달할 수는 없었음"
    - 파이썬에서는 함수도 값(value)처럼 다룰 수 있기 때문에 생긴 용어임
        - 파이썬, 자바스크립트, 리스프 같은 언어는,“함수도 값처럼 다뤄질 수 있다”는 개념을 도입
    - 즉, 숫자나 문자열처럼 함수를 변수에 담거나, 다른 함수의 인자로 전달하거나, 함수의 반환값으로 사용할 수 있음
    - 일급함수의 특징
        - 변수에 저장될 수 있고,
        - 다른 함수의 인자로 전달되고,
        - 함수의 반환값이 될 수 있다면
- 예: map(), filter(), sorted(), any(), all() 등
- 람다 함수는 “짧고 간단한 일회성 연산”에는 유용하지만, “복잡한 계산·디버깅·문서화”에는 def 함수가 훨씬 낫습니다.

### ‘람다(lambda)’의 어원
- Lambda(람다): 그리스 알파벳의 11번째 문자(λ)
- 이 단어는 1930년대 수학자 알론조 처치(Alonzo Church)가 만든 “람다 계산법(Lambda Calculus)”이라는 수학적 모델에서 유래
    - 검색해보세요: '알론조 처치(Alonzo Church)는 왜 람다 라고 지칭했을까?'
- 람다 계산법(Lambda Calculus): → “함수를 수학적으로 표현하고 계산하는 아주 기초적인 체계”
- 이 계산법에서 함수를 정의할 때 ‘λ(람다)’ 기호를 사용
- 수학식으로: λx. x + 1  ==> 파이썬으로  lambda x: x + 1
- 다음과 같이 함수를 사용해야 하지만 간단하게 기술하고자함.
    - 수학처럼 한 줄로 표현할 수 있는 익명 함수(anonymous function)가 필요했고,  “이름 없는 짧은 함수”, 즉 익명 함수로 람다사용

```python
# 일반 함수
def add(x, y):
    return x + y

print(add(3,4))


# 람다 함수 (한 줄)
add_lambda = lambda x, y: x + y
print(add_lambda(2, 3))  # 결과: 5
```

#### [2-1] map() +lambda

```python
###########################3
## 람다없을때
###########################
# 1. 함수 정의
def multiply_two(x):
    return x * 2

# 2. map에 함수 이름만 전달
nums = [1, 2, 3, 4, 5]
result = list(map(multiply_two, nums))

print(result)
```

```python
#########################################
# 리스트의 모든 값을 일정 규칙으로 변환
# map() + lambda
# 리스트의 모든 값을 일정 규칙으로 변환
#########################################

nums = [1, 2, 3, 4, 5]

# 각 숫자에 2를 곱하기
result = list(map(lambda x: x * 2, nums))
print(result)
```

#### [2-3] sorted() + lambda
- 정렬 기준 함수를 직접 지정

```python
###########################
## 람다x
###########################

# 길이를 반환하는 함수 정의
def get_length(word):
    return len(word)

words = ["apple", "banana", "kiwi", "strawberry"]

# key에 함수 이름만 전달
result = sorted(words, key=get_length)
print(result)
```

```python
##################################
## sorted() + lambda
## 정렬 기준 함수를 직접 지정
######################################
words = ["apple", "banana", "kiwi", "strawberry"]

# 글자 길이로 정렬
result = sorted(words, key=lambda w: len(w))
print(result)
```
